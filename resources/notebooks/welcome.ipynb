## sharing files
The sparkplayground vhdx has a NTFS partition, and file sharing is enabled.
From your windows host, you can access the files, for instance like below

```
NET USE P: \\sparkplayground.mshome.net\shared sharesparkadminpassword /USER:sparkadmin /PERSISTENT:no
```
