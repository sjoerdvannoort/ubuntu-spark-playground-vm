```python
%%sparksql
CREATE TABLE fact_sales 
AS
SELECT so.CustomerID, so.OrderID, so.StoreID
, sd.ProductID
, so.OrderDate
, sd.Quantity
, sd.TotalPrice
FROM sales_demo.sales_orders as so
JOIN sales_demo.sales_details as sd USING (OrderID)
```

```python
%%sparksql
SHOW TABLES
```

```python
%%sparksql
DESCRIBE TABLE EXTENDED fact_sales
```

```python
%%sparksql
SELECT *
FROM fact_sales
```

```python
%%sparksql
DESCRIBE HISTORY fact_sales
```

```python
%%sparksql
DROP TABLE fact_sales
```
