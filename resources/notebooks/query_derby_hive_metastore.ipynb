```python
# JDBC connection properties
jdbc_url = "jdbc:derby://localhost:1527//opt/sparkplayground/metastore/hive_derby"
jdbc_driver = "org.apache.derby.jdbc.ClientDriver"

table_name = "SYS.SYSTABLES"
# Read data from the JDBC source
df = spark.read.format("jdbc") \
    .option("url", jdbc_url) \
    .option("dbtable", table_name) \
    .option("driver", jdbc_driver) \
    .load()

df.where("TABLETYPE NOT IN('S')").show()
```
